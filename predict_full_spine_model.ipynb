```python
## Import
from __future__ import print_function
from utils.patient import Patient

import torch

from unet.unet_model import UNet
from utils.patient import Patient

import glob
import os

print("CUDA Status: " + str(torch.cuda.is_available()))
```

```python
## Load and predict
n_channels = 1
n_classes = 14
net = UNet(n_channels=n_channels, n_classes=n_classes)
net.to(device='cuda' if torch.cuda.is_available() else 'cpu')

best_epoch = [6,55,60,18,63]

for partition in range(5):
    modelpath = "D:\\autoseg\\nets\\unet-spine-save" + str(partition) + "_" + str(best_epoch[partition]) + ".pth"
    print("Loading net at: " + modelpath)
    
    net.load_state_dict(torch.load(modelpath, map_location='cuda' if torch.cuda.is_available() else 'cpu'))
    
#     image_path = "Z:/projects/autoseg/full-img"
#     partn_path = "Z:/projects/autoseg/full-pt" + str(partition)
    image_path = "D:/autoseg/full-img"
    partn_path = "D:/autoseg/full-spine-pt" + str(partition)
    if not os.path.exists(partn_path):
        os.mkdir(partn_path)
    
    
    for filename in glob.glob(image_path+"/CTCAC*"):
        fnspl = filename.split("\\")
        patname = fnspl[-1].replace(".nii.gz", "")
        print(patname)
        pat_path = filename
        seg_path = partn_path + "/" + patname + ".nii.gz"
        
#         print(pat_path)
#         print(seg_path)

        patient = Patient(pat_path,data_device='cuda' if torch.cuda.is_available() else 'cpu')

        patient.predict(net,seg_path,batch_size=10)
```
